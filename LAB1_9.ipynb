```python
import random
from collections import defaultdict

class Agent:
    def __init__(self, grid_size=(5, 5)):
        self.grid_size = grid_size
        self.position = [0, 0]  
        self.dirt_map = self.generate_dirt_map() 
    
    def generate_dirt_map(self):
        return {(x, y): random.choice([0, 1]) for x in range(self.grid_size[0]) for y in range(self.grid_size[1])}
    
    def perceive_environment(self):
        return self.dirt_map[tuple(self.position)]
    
    def move(self, direction):
        x, y = self.position
        if direction == 'up' and x > 0:
            self.position[0] -= 1
        elif direction == 'down' and x < self.grid_size[0] - 1:
            self.position[0] += 1
        elif direction == 'left' and y > 0:
            self.position[1] -= 1
        elif direction == 'right' and y < self.grid_size[1] - 1:
            self.position[1] += 1
    
    def clean(self):
        if self.perceive_environment():
            self.dirt_map[tuple(self.position)] = 0 
            return "Cleaning..."
        return "Already clean."
    
    def run(self):
        directions = ['up', 'down', 'left', 'right']
        for _ in range(10): 
            print(f"Current Position: {self.position}")
            print(self.clean())
            direction = random.choice(directions)
            self.move(direction)
            print(f"Moved {direction}\n")
        print("Cleaning session complete.")

agent = Agent()
agent.run()
```
